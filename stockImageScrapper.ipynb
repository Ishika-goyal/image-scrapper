```python
s = Service(r"C:\Program Files (x86)\chromedriver.exe")
browser = webdriver.Chrome(service=s)
```

```python
browser.get(link)
```

```python
soup=BeautifulSoup(browser.page_source,'html.parser')
```

```python
soup.prettify
```

```python
soup.find_all('img')
```

```python
data=[]
for sp in soup.find_all('div',class_='container'):
    #print(sp)
    link=sp.find('img').get('src')
    tags=sp.find('div',class_='tags').text
    likes=(sp.find_all('span')[1].text)[:-6]
    comments=(sp.find_all('span')[2].text)[:-9]
    tags=''.join(tags)
    data.append([link,likes,comments,tags])
    
```

```python
data[10]
```

```python
df=pd.DataFrame(data,columns=['link','likes','comments','tags'])
```
